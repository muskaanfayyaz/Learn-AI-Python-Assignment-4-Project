## **Guess the Number Game Python Project (computer)**


```python
def guess_the_number_user_picks():
    print("Think of a number between 1 and 100, and I will try to guess it.")

    low = 1
    high = 100
    guess = None
    attempts = 0

    while True:
        # Computer guesses the middle value
        guess = (low + high) // 2
        attempts += 1

        # Ask user for feedback on the guess
        print(f"My guess is: {guess}")
        feedback = input("Is it too high, too low, or correct? (high/low/correct): ").lower()

        if feedback == "correct":
            print(f"I guessed your number {guess} in {attempts} attempts! Yay!")
            break
        elif feedback == "high":
            high = guess - 1  # The number must be lower
        elif feedback == "low":
            low = guess + 1  # The number must be higher
        else:
            print("Invalid input! Please respond with 'high', 'low', or 'correct'.")

# Call the function to start the game
guess_the_number_user_picks()
```

## **Guess the Number Game Python Project (user)**


```python
import random

# Function to start the Guess the Number game
def guess_the_number():
    print("Welcome to Guess the Number!")
    print("I am thinking of a number between 1 and 100.")

    # Random number generated by the computer
    number_to_guess = random.randint(1, 100)

    # Initialize the number of guesses taken
    guess_count = 0

    while True:
        # Get user's guess
        user_guess = int(input("Enter your guess: "))
        guess_count += 1

        # Check the guess
        if user_guess < number_to_guess:
            print("Too low! Try again.")
        elif user_guess > number_to_guess:
            print("Too high! Try again.")
        else:
            print(f"Congratulations! You guessed the number in {guess_count} tries.")
            break  # Exit the loop when the user guesses correctly

# Call the function to start the game
guess_the_number()
```

## **Rock, paper, scissors Python Project**

```python
import random

# Function to determine the winner
def determine_winner(user_choice, computer_choice):
    if user_choice == computer_choice:
        return "It's a tie!"
    elif user_choice == 'rock':
        return "You win!" if computer_choice == 'scissors' else "You lose!"
    elif user_choice == 'scissors':
        return "You win!" if computer_choice == 'paper' else "You lose!"
    elif user_choice == 'paper':
        return "You win!" if computer_choice == 'rock' else "You lose!"
    else:
        return "Invalid choice!"

# Function to play the game
def play_rock_paper_scissors():
    choices = ['rock', 'paper', 'scissors']

    while True:
        # Get the user's choice
        user_choice = input("Enter 'rock', 'paper', or 'scissors' (or 'quit' to stop playing): ").lower()

        if user_choice == 'quit':
            print("Thanks for playing!")
            break
        if user_choice not in choices:
            print("Invalid choice. Please choose 'rock', 'paper', or 'scissors'.")
            continue

        # Get the computer's choice
        computer_choice = random.choice(choices)
        print(f"Computer chose: {computer_choice}")

        # Determine and display the winner
        result = determine_winner(user_choice, computer_choice)
        print(result)

# Call the function to start the game
play_rock_paper_scissors()
```

## **Hangman Python Project**

```python
import random

# List of words to choose from
word_list = ['python', 'hangman', 'programming', 'developer', 'computer', 'keyboard']

# Function to choose a random word
def get_random_word():
    return random.choice(word_list)

# Function to display the current state of the word
def display_word(word, guessed_letters):
    display = ''
    for letter in word:
        if letter in guessed_letters:
            display += letter + ' '
        else:
            display += '_ '
    return display.strip()

# Function to play Hangman
def play_hangman():
    word = get_random_word()  # Choose a random word
    guessed_letters = []  # List to store guessed letters
    attempts = 6  # Number of allowed incorrect guesses
    guessed_word = False  # Track if the word is guessed correctly

    print("Welcome to Hangman!")
    print("Try to guess the word, one letter at a time.")

    while attempts > 0 and not guessed_word:
        print("\nCurrent word:", display_word(word, guessed_letters))
        print(f"You have {attempts} attempts remaining.")

        # Get the user's guess
        guess = input("Enter a letter: ").lower()

        # Check for valid input
        if len(guess) != 1 or not guess.isalpha():
            print("Please enter a single valid letter.")
            continue

        # Check if the letter has already been guessed
        if guess in guessed_letters:
            print("You've already guessed that letter. Try again.")
            continue

        guessed_letters.append(guess)

        # Check if the guessed letter is in the word
        if guess in word:
            print(f"Good job! {guess} is in the word.")
        else:
            attempts -= 1
            print(f"Sorry! {guess} is not in the word.")

        # Check if the user has guessed the word correctly
        if all(letter in guessed_letters for letter in word):
            guessed_word = True

    # End of the game
    if guessed_word:
        print(f"\nCongratulations! You've guessed the word: {word}")
    else:
        print(f"\nYou've run out of attempts! The word was: {word}")

# Call the function to start the game
play_hangman()
```

## **Countdown Timer Python Project**

```python
import time
from IPython.display import Audio, display

# Function to play the alarm sound from your GitHub repo
def play_sound():
    sound_url = 'https://raw.githubusercontent.com/muskaanfayyaz/Learn-AI-Python-Assignment-4-Project/main/alarm_sound.mp3'
    display(Audio(url=sound_url, autoplay=True))

# Countdown timer function
def countdown_timer(seconds):
    while seconds:
        mins, secs = divmod(seconds, 60)
        timeformat = '{:02d}:{:02d}'.format(mins, secs)
        print(timeformat, end='\r')
        time.sleep(1)
        seconds -= 1

    print("00:00\nTime's up!")
    play_sound()

# Get user input and start the timer
def get_user_input():
    try:
        minutes = int(input("Enter minutes: "))
        seconds = int(input("Enter seconds: "))
        total_seconds = minutes * 60 + seconds
        countdown_timer(total_seconds)
    except ValueError:
        print("Please enter valid numbers.")

# Run the timer
get_user_input()
```
